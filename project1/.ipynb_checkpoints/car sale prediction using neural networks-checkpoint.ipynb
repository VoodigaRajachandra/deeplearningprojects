### converting categorical to numerical

```python
data.nunique()
```

```python
print(data.dtypes)
```

```python
num_data=data.drop(['Name', 'Location', 'Fuel_Type', 'Transmission', 'Owner_Type', 'Mileage','Engine', 'Power'],axis=1)
```

```python
num_data.head()
```

```python
cate_data = data.filter(['Location', 'Fuel_Type', 'Transmission', 'Owner_Type', 'Mileage','Engine', 'Power'], axis=1)
cate_data.head()
```

### one hot encoding

```python
cate_data_one_hot=pd.get_dummies(cate_data,drop_first=True)
```

```python
cate_data_one_hot.head()
```

```python
complete_data=pd.concat([num_data,cate_data_one_hot],axis=1)
```
